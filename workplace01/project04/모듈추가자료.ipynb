```python
# b64/decoder.py
import base64


def base64_to_str(x):
    """base64 표현을 문자열로 변환함

    b64decode()의 반환값은 bytes 타입이므로
    decode()로 문자열로 바꾼 뒤 변환함
    """
    return base64.b64decode(x).decode('utf-8')
```

### 인클로징 스코프 ── 현재 로컬 스코프의 1계층 바깥의 스코프

```python
# x
def f():
    x = 'x'
    def g():
        print(x)  # 인클로징 스코프 안의 변수 x를 참조
    g()
```

```python
def f():
    x = 'x'
    def g():
        nonlocal x  # x가 로컬 변수가 아님을 선언
        x = 1
        print(x)
    g()
    print(x)
```

#### 클로저 ── 작성 시의 환경을 기억하는 함수 객체

```python
def counter():
    count = 0
    def _increment():
        nonlocal count
        count += 1
        return count
    return _increment
```

```python
# 카운터 작성
counter1 = counter()
counter1
```

```python
# _increment의 바깥쪽에서 정의되어 있던
# count의 참조를 계속 유지하고 있음
counter1()
```

```python
counter1()
```

```python
# 다른 카운터 작성
# counter1이 참조하고 있는 count에는 영향이 없음
counter2 = counter()
counter2()
```
